# Configuration de l'environnement de travail

La premire tape sera de configurer l'environnement de travail. Pour cela, nous allons installer le logiciel Thonny qui nous permettra d'crire et d'excuter nos programmes en Python.

## Installation de Thonny
Pour installer Thonny, il faudra le tlcharger

Veuillez suivre les tapes suivantes :
1. Ouvrez un navigateur web.
2. Allez sur le site web de Thonny : https://thonny.org/
3. Tlchargez la version de Thonny qui correspond  votre systme d'exploitation soit Windows, Mac ou Linux.
   - Prenez le premier lien. Au moment d'crire ces lignes, j'utilisais la version 4.1.1 de Thonny.
  
   - 
4. Atteignez le dossier des tlchargements de votre ordinateur.
5. Double-cliquez sur le fichier tlcharg pour lancer l'installation.
6. Suivez les instructions d'installation. Gnralement, ce sera de cliquer sur le bouton `Suivant` jusqu' ce que l'installation soit termine.

Vous devriez maintenant avoir Thonny d'install sur votre ordinateur. Pour valider, allez dans le menu dmarrer de Windows et cherchez Thonny.



## Testez Thonny
Pour tester Thonny, nous allons crire un programme qui affiche `Bonjour le monde!`  l'cran.

Excutez les tapes suivantes :
1. Ouvrez Thonny.
2. Lors de la premire excution, il demandera de choisir la langue. Choisissez `Franais`.
3. Tapez le code suivant dans la fentre de Thonny :
   ```python
   print("Bonjour le monde!")
   ```
4. Cliquez sur le bouton `Excuter` ou appuyez sur la touche `F5` de votre clavier.
   - Le bouton est le petit triangle vers qui ressemble  un bouton `Play` sur un lecteur de musique.
5. Dans la console, vous devriez voir le message `Bonjour le monde!` s'afficher.

# Tour d'horizon de Thonny
Thonny est un environnement de travail qui est trs simple  utiliser. Il est donc parfait pour les dbutants. Dans cette section, nous allons faire un tour d'horizon de Thonny.

## L'interface de Thonny
Lorsque vous ouvrez Thonny, vous devriez voir une fentre qui ressemble  ceci :



Comme la majorit des applications Windows, on retrouve une barre de menu en haut de la fentre. Cette barre de menu contient les menus suivants :
- `Fichier` : Contient les options pour crer, ouvrir, enregistrer et imprimer un fichier.
- `diter` : Contient les options pour copier, couper, coller et annuler.
- `Affichage` : Contient les options pour afficher ou masquer les diffrents panneaux de Thonny.
- `Excuter` : Contient les options pour excuter le code.
- `Outils` : Contient les options pour configurer Thonny.
- `Aide` : Contient les options pour obtenir de l'aide sur Thonny.

En dessous de la barre de menu, on retrouve une barre d'outils. Cette barre d'outils contient les boutons suivants :
- `Nouveau` : Permet de crer un nouveau fichier.
- `Ouvrir` : Permet d'ouvrir un fichier existant.
- `Enregistrer` : Permet d'enregistrer le fichier courant.
- `Excuter` : Permet d'excuter le code.
- `Dboguer` : Permet de dboguer le code.
  - Quelques boutons en lien avec le dbogage sont griss. Nous verrons en cours de session comment les utiliser.
- `Arrter` : Permet d'arrter l'excution du code.

En dessous de la barre d'outils, on retrouve la fentre principale de Thonny. Lors de la premire excution, cette fentre est divise en deux parties :
- La partie du haut est l'diteur de code. C'est ici que vous allez crire votre code.
- La partie du bas est la console. C'est ici que vous allez voir le rsultat de l'excution de votre code.

## Les panneaux de Thonny
Thonny possde plusieurs panneaux qui permettent d'afficher des informations supplmentaires. Ces panneaux se retrouvent dans le menu `Affichage`. Pour l'instant, nous allons nous concentrer que sur le fonctionnement de Thonny. Nous verrons les panneaux plus tard dans le cours.



Pour afficher un panneau, il suffit de cliquer sur le nom du panneau dans le menu `Affichage`. Pour masquer un panneau, il suffit de cliquer sur le nom du panneau dans le menu `Affichage` ou de cliquer sur le bouton `X` dans le coin suprieur droit du panneau.

Par exemple, si l'on clique par accident sur le `X` de la console, on pourra la faire rapparatre en cliquant sur `Affichage` puis `Console`.
